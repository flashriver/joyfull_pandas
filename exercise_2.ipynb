### (3) 移动平均法是一种时间序列的常见平滑方式，可分为 k 期移动平均和 k期中心移动平均，都使用了某一时刻及其周围的数据对该时刻的数据进行平滑修正。
设原序列为 $x_1 ,...,x_n$ ，对于 $x_t$ 的 $k$ 期移动平均修正 $\tilde{x_t}$ 为
$$\frac{∑^{k−1}_{i=0}x_{t−i}}{k}$$
，对于 k 期中心移动平均修正为

$$
\tilde{x_t} =
\begin{cases}
\frac{1}{k}(\frac{1}{2}x_{t-\frac{k}{2}}+x_{t-\frac{k}{2}+1}+...+x_{t}+...+x_{t+\frac{k}{2}-1}+\frac{1}{2}x_{t+\frac{k}{2}}) & \text{ , k 为偶数 } \\
\frac{1}{k}(x_{t-\frac{k-1}{2}}+x_{t-\frac{k-1}{2}+1}+...+x_{t}+...+x_{t+\frac{k-1}{2}-1}+x_{t+\frac{k-1}{2}}) & \text{ , k 为奇数 } 
\end{cases} 
$$

(a) 求原序列的 5 期移动平均序列。   
(b) 求原序列的 5 期与 6 期中心移动平均序列。

```python
# (a)
d[['Date','Temp']].set_index('Date').rolling(window=5).mean()
```

```python
# (b) 5 期中心移动平均序列
d3 = d[['Date','Temp']].set_index('Date')
d3.rolling(window=5).apply(lambda x:x.sum()/5).shift(-2)
```

```python
# 6 期移动平均序列
d3.rolling(window=7).apply(lambda x:(x.sum()-x[0]/2-x[6]/2)/6).shift(-3)
```

```python
# (b) 7 期中心移动平均序列
d3.rolling(window=7).apply(lambda x:x.sum()/7).shift(-3)
```

```python
# 8 期移动平均序列
d3.rolling(window=9).apply(lambda x:(x.sum()-x[0]/2-x[8]/2)/8).shift(-4)
```

```python
# 将 lambda 函数显式定义出来
def mvmn(df,k):
    if k%2==1:
        return df.rolling(window=k).apply(lambda x:x.sum()/k).shift(-(k//2))
    elif k%2==0:
        return df.rolling(window=k+1).apply(lambda x:(x.sum()-x[0]/2-x[k]/2)/k).shift(-(k//2))
```

```python
# 5 期中心移动平均序列
mvmn(d3,5).equals(d3.rolling(window=5).apply(lambda x:x.sum()/5).shift(-2))
```

```python
# 6
mvmn(d3,6).equals(d3.rolling(window=7).apply(lambda x:(x.sum()-x[0]/2-x[6]/2)/6).shift(-3))#d3.rolling(window=7).apply(lambda x:x.sum()/6).shift(-3))
```

```python
# 7
mvmn(d3,7).equals(d3.rolling(window=7).apply(lambda x:x.sum()/7).shift(-3))
```

```python
# 8
mvmn(d3,8).equals(d3.rolling(window=9).apply(lambda x:(x.sum()-x[0]/2-x[8]/2)/8).shift(-4))
```
